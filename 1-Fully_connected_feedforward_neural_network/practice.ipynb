```python
import numpy as np
```

```python
def activation(x):
    return 0 if x < 0.5 else 1
```

```python
# функция которая пропускает через нейронную сеть входной сигнал
# на выходе смотрим образовалась ли симпатия
```

```python
# house - дом
# rock - рок
# attr - привлекательность
# параметры принимают 0 или 1

def go(house, rock, attr):
    # формируем вектор входного сигнала
    x = np.array([house, rock, attr])
    # веса 1 и 2 нейрона скрытого слоя
    w11 = [0.3, 0.3, 0]
    w12 = [0.4, -0.5, 1]
    # связи между слоями нейронов
    weight1 = np.array([w11, w12]) # матрица 2*3
    weight2 = np.array([-1, 1]) # вектор 1*2

    # сумма на входах нейронов скрытого слоя
    sum_hidden = np.dot(weight1, x)
    print("Значения сумм на нейронах скрытого слоя: "+str(sum_hidden))

    out_hidden = np.array([activation(x) for x in sum_hidden])
    print("Значения на выходах нейронов скрытого слоя: "+str(out_hidden))

    sum_end = np.dot(weight2, out_hidden)
    y = activation(sum_end)
    print("Выходное значение НС: "+str(y))

    return y
```

```python
# есть дом, не любит рок, привлекателен
house = 1
rock = 0
attr = 1
res = go(house, rock, attr)

if res == 1:
    print("Ты мне нравишься")
else:
    print("Ты мой самый лучший друг")
```
