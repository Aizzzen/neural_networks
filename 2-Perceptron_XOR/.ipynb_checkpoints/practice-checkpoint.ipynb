```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# моделируем образы, их будет по 5 для каждого класса
```

```python
N = 5
```

```python
x1 = np.random.random(N) # случайные величины по одной оси
x2 = x1 + [np.random.randint(10)/10 for i in range(N)] # + случайное отклонение
C1 = [x1, x2] # т.о. точки будут лежать выше прямой
```

```python
x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 # чтобы точка ТОЧНО была ниже нуля
C2 = [x1, x2]
```

```python
# прямая под 45*
f = [0, 1]
```

```python
# веса
w = np.array([-0.3, 0.3])
```

```python
for i in range(N):
    # перебираем все образы для 2 класса
    x = np.array([C2[0][i], C2[1][i]])
    # вычисляем выходное значение
    y = np.dot(w, x)
    
    if y >= 0:
        print("Класс C1")
    else:
        print("Класс C2")
```

```python
plt.scatter(C1[0][:], C1[1][:], s=10, c='red')
plt.scatter(C2[0][:], C2[1][:], s=10, c='blue')
plt.plot(f)
plt.grid(True)
plt.show()
```

```python
# смещение
```

```python
N = 5
b = 3
```

```python
x1 = np.random.random(N) # случайные величины по одной оси
x2 = x1 + [np.random.randint(10)/10 for i in range(N)] + b # + случайное отклонение
C1 = [x1, x2] # т.о. точки будут лежать выше прямой
```

```python
x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 + b # чтобы точка ТОЧНО была ниже нуля
C2 = [x1, x2]
```

```python
# прямая под 45*
f = [0 + b, 1 + b]
```

```python
# веса
w2 = 0.5
w3 = -b * w2
w = np.array([-w2, w2, w3])
```

```python
for i in range(N):
    # перебираем все образы для 2 класса
    x = np.array([C1[0][i], C1[1][i], 1])
    # вычисляем выходное значение
    y = np.dot(w, x)
    
    if y >= 0:
        print("Класс C1")
    else:
        print("Класс C2")
```

```python
plt.scatter(C1[0][:], C1[1][:], s=10, c='red')
plt.scatter(C2[0][:], C2[1][:], s=10, c='blue')
plt.plot(f)
plt.grid(True)
plt.show()
```

```python
# как видно, прямая смещена по вертикали на 3 единицы
# образы тоже смещены
# при этом происходит верная классификация
```
